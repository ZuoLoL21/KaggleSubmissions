### Checking if the datasets have the same categories

```python
import pandas as pd

train_df = pd.read_csv(
        'C:/Users/zuolo/Documents/_Personal/MLStuff/Kaggle/projects/4-store-sales-time-series-forecasting/inputs/raw/train.csv')
df_counts = train_df.groupby(['store_nbr', 'date']).size().reset_index(name='Count')

print("All same for train? : ", df_counts['Count'].nunique() == 1)
print(df_counts['Count'][0])
tmp_numb = df_counts['Count'][0]

test_df = pd.read_csv(
        'C:/Users/zuolo/Documents/_Personal/MLStuff/Kaggle/projects/4-store-sales-time-series-forecasting/inputs/raw/test.csv')
df_counts = test_df.groupby(['store_nbr', 'date']).size().reset_index(name='Count')

print("All same for test?  : ", df_counts['Count'].nunique() == 1)
print(df_counts['Count'][0])

print("All the same categories? : ", df_counts['Count'][0] == tmp_numb)
```

```python
from typing import Iterable
import pandas as pd


def _left_join(df1, df2, axis_name: Iterable[str] = 'index'):
    return df1.merge(df2, on=axis_name, how='left')


def _right_join(df1, df2, axis_name: Iterable[str] = 'index'):
    return df2.merge(df1, on=axis_name, how='right')


train_df = pd.read_csv(
        "C:/Users/zuolo/Documents/_Personal/MLStuff/Kaggle/projects/4-store-sales-time-series-forecasting/inputs/transformed/train.csv",
        delimiter=",")
oil_df = pd.read_csv(
        "C:/Users/zuolo/Documents/_Personal/MLStuff/Kaggle/projects/4-store-sales-time-series-forecasting/inputs/transformed/oil.csv",
        delimiter=",")
_right_join(train_df, oil_df, axis_name='date').head()
```
